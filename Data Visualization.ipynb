```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline #to allow to see graphs inside the jupyter notebook
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL METHOD
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.show() #just to print in notebook other than jupyter
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oiented method
fig = plt.figure()#fig is object
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3]) # arguments in order from i % : left,bottom,width,height
axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes=plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

# FIG SIZE and DPI

```python
fig = plt.figure(figsize=(3,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#save a figure
fig.savefig('my_picture.jpg')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x-squared')
ax.plot(x,x**3,label='x-cubed')
ax.legend()
ax.legend(loc=(0.1,0.1))
```

# PLOT Apperance

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=1,alpha=2,ls='-',marker='o',markersize=15,markerfacecolor='green'
        ,markeredgewidth=3,markeredgecolor='black')
#we can also use RGB Hex codes in place of basic color names #alpha is for tranparency
#ls= linestyle lw=linewidth
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=8,alpha=2,ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
