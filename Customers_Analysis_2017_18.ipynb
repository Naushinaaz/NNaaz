```python
#Importing the input CSV file
import pandas as pd
import numpy as np
import os
df= pd.read_csv("./All_Plants_Input.csv",encoding="latin-1")
#with open(filename, encoding="latin-1") as df:
#cleaning the dataframe
#cleaning the dataframe
df.isnull().values.any()
df.fillna('X', inplace=True)
df = df.convert_objects(convert_numeric=True)
df['Cust_ID'] = df['Cust_ID'].astype(str)
df['Plant_ID']=df['Plant_ID'].astype(str)
pd.options.display.float_format = '{:.2f}'.format
df['Mar_Diff']=df['PM_2018']-df['PM_2017']
df.head(20)
```

```python
df.dtypes
```

```python
df.sort_values('NACS_Category',ascending =True)
df
```

```python
x=df.groupby(['NACS_Category'])['PM_2017'].count()
x
```

```python
y=df.groupby(['NACS_Category'])['PM_2018'].count()
y
```

```python
df_ref=df.groupby(['NACS_Category']).sum()

df_ref
```

```python
Cus_data=pd.DataFrame({'Cust_2017':x.values,'Cust_2018':y.values},index=x.keys())
Cus_data.dtypes
```

```python
Cus_data
```

```python
final_df=pd.merge(df_ref,Cus_data,how='inner', left_index=True,right_index=True)
final_df
```

```python
os.makedirs('Customer_Ratio')
writer = pd.ExcelWriter('Customer_Ratio/All_Plants_CustomerRatio.xlsx')
final_df1= final_df.applymap(lambda x: round(x,2) if isinstance (x,(int,float)) else x)
final_df1.to_excel(writer,'All_Plants_CustomerRatio')
writer.save()
```
