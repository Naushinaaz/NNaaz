```python
import pandas as pd
import numpy as np
df = pd.read_csv("./File_for_python_Test2.csv")
```

```python
df.fillna('X', inplace=True)
```

```python
df['Sold-to Customer'] = df['Sold-to Customer'].astype(str)
df = df.convert_objects(convert_numeric=True)
```

```python
df.dtypes
```

```python
df.groupby(['FineLines']).mean().sort_values('PM_2017', ascending=False)
```

```python
df['MarDiff']=df['PM_2018']-df['PM_2017']
```

```python
MD_finelines=df.groupby(["FineLines"]).sum().sort_values("MarDiff", ascending=True)
```

```python
MD_finelines
```

```python
margin_out=MD_finelines[MD_finelines['MarDiff']<0].sort_values('MarDiff',ascending =True)
```
