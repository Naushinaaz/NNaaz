```python
#Importing the input CSV file
import pandas as pd
import numpy as np
import os
df= pd.read_csv('Zootopia_All_plants_copy.csv')
df_map= pd.read_csv('Fineline_CAT_mapping.csv',index_col='NACS_Category')
#cleaning the dataframe

df.head(20)
```

```python
import matplotlib.pyplot as plt
```

```python
from functools import reduce
```

```python
def percentage(percent, whole):
    return (percent/whole) * 100.0
```

```python
df['Cust_ID'] = df['Cust_ID'].astype(str)
pd.options.display.float_format = '{:.2f}'.format
df['Mar_Diff']=df['PM_2018']-df['PM_2017']
df.dtypes
```

```python
#picking valid mar_diff i.e. removing nan for mar_diff
df = df[np.isfinite(df['Mar_Diff'])]
len(df)
```

```python
df_both_years=df.dropna(axis=0,subset=['PM_2017','PM_2018'],inplace=False) #customers who bought in 2017and 2018
df_both_years
```

```python
df_C_PM=pd.DataFrame(df_both_years.groupby(['NACS_Category'])['PM_2018'].count())
df_C_PM.rename(columns={'PM_2018': 'Flag_C_PM'}, inplace=True)
len(df_C_PM) #Flag_C_PM
```

```python
#Margin_out dataframe has declining margins
margin_out=df_both_years[df_both_years['Mar_Diff']<0].sort_values('NACS_Category',ascending =True)
df_D_PM=pd.DataFrame(margin_out.groupby(['NACS_Category'])['Cust_ID'].count())
df_D_PM.rename(columns={'Cust_ID': 'Flag_D_PM'}, inplace=True)
len(df_D_PM) #Flag_D_PM
```

```python
#increase in margin
margin_inc=df_both_years[df_both_years['Mar_Diff']>0].sort_values('NACS_Category',ascending =True)
df_I_PM=pd.DataFrame(margin_inc.groupby(['NACS_Category'])['Cust_ID'].count())
df_I_PM.rename(columns={'Cust_ID': 'Flag_I_PM'}, inplace=True)
len(df_I_PM) #Flag_I_PM
```

```python
#include another flag to show margin difference equal to zero
margin_inc=df_both_years[df_both_years['Mar_Diff']==0].sort_values('NACS_Category',ascending =True)
df_S_PM=pd.DataFrame(margin_inc.groupby(['NACS_Category'])['Cust_ID'].count())
df_S_PM.rename(columns={'Cust_ID': 'Flag_S_PM'}, inplace=True)
len(df_S_PM)
```

```python
dfs = [df_D_PM, df_I_PM, df_C_PM,df_S_PM]
df_temp1 = reduce(lambda left,right: pd.merge(left,right,how='outer',left_index=True,right_index=True), dfs)
df_temp1['DM_factor']=percentage(df_temp1['Flag_D_PM'],df_temp1['Flag_C_PM'])
df_temp1['IM_factor']=percentage(df_temp1['Flag_I_PM'],df_temp1['Flag_C_PM'])
df_temp1['SM_factor']=percentage(df_temp1['Flag_S_PM'],df_temp1['Flag_C_PM'])
df_temp1.head()
df_temp1.dtypes
```

```python
conditions=[(df_temp1['DM_factor'] != df_temp1['IM_factor'])
                                   & (df_temp1['IM_factor']!= df_temp1['SM_factor'])
                                    & (df_temp1['DM_factor']!= df_temp1['SM_factor']),
            (df_temp1['DM_factor'] == df_temp1['IM_factor'])
                                   & (df_temp1['IM_factor']!= df_temp1['SM_factor'])
                                    & (df_temp1['DM_factor']!= df_temp1['SM_factor']),
            (df_temp1['DM_factor'] != df_temp1['IM_factor'])
                                   & (df_temp1['IM_factor']== df_temp1['SM_factor'])
                                    & (df_temp1['DM_factor']!= df_temp1['SM_factor']),
            (df_temp1['DM_factor'] != df_temp1['IM_factor'])
                                   & (df_temp1['IM_factor']!= df_temp1['SM_factor'])
                                    & (df_temp1['DM_factor']== df_temp1['SM_factor']),
            (df_temp1['DM_factor'] == df_temp1['IM_factor'])
                                   & (df_temp1['IM_factor']== df_temp1['SM_factor'])
                                    & (df_temp1['DM_factor']== df_temp1['SM_factor'])]
choices=[df_temp1[['IM_factor', 'DM_factor','SM_factor']].idxmax(axis=1),'IM_factor','IM_factor','DM_factor',
         'IM_factor']
            
df_temp1['Final_factor']=np.select(conditions, choices, default=np.nan)
```

```python
df_CP_factor=df.groupby(['NACS_Category'])['CNP_2017','CNP_2018'].count()
df_CP_factor['CP_factor']=(df_CP_factor['CNP_2018']/df_CP_factor['CNP_2017'])*100
df_CP_factor.rename(columns={'CNP_2017': 'CustNum_2017','CNP_2018': 'CustNum_2018'}, inplace=True)
df_CP_factor.head()
```

```python
print (len(df_CP_factor))
```

```python
df_CS_factor=df.groupby(['NACS_Category'])['CNP_2017','CNP_2018'].sum()
df_CS_factor['CS_factor']=(df_CS_factor['CNP_2018']/df_CS_factor['CNP_2017'])*100
df_CS_factor.head()
```

```python
df_temp2=df_CP_factor.join(df_CS_factor)
df_temp2.head()
```

```python
print ("No of records in temp2 :- ", len(df_temp2))
print ("No of records in temp1 :- ", len(df_temp1))
df_param=pd.merge(df_temp2,df_temp1,how='outer',left_index=True,right_index=True)
df_param.head()
```

```python
sales_inc=df_param['CS_factor'] >100
sales_same=df_param['CS_factor']==100
sales_dec=df_param['CS_factor'] <100
mar_inc=df_param['Final_factor']=='IM_factor'
mar_dec=df_param['Final_factor']=='DM_factor'
mar_same=df_param['Final_factor']=='SM_factor'
```

```python
conditions = [
    (mar_inc & sales_inc),
    (mar_inc & sales_same),
    (mar_inc & sales_dec),(mar_same & sales_inc),(mar_same & sales_dec),(mar_same & sales_same),
    (mar_dec & sales_inc),(mar_dec & sales_same),(mar_dec & sales_dec)]
choices=['1.A+','2.A','3.A-','4.B+', '5.B','6.B-','7.C+','8.C','9.C-']
df_param['label']= np.select(conditions, choices, default='')
```

```python
len(df_param)
```

```python
df_param.reset_index(inplace=True)
df_map.reset_index(inplace=True)
df_mapped=pd.merge(df_param,df_map,how='inner',on='NACS_Category')
df_mapped
```

```python
df_final= pd.DataFrame(df_mapped.groupby(['Categories','label'])['NACS_Category'].count())
df_final.head()
```

```python
writer = pd.ExcelWriter('Final_out/Margin_Analysis1.xlsx')
df_TEMP1=df_mapped.applymap(lambda x: round(x,2) if isinstance (x,(int,float)) else x)
df_param1=df_param.applymap(lambda x: round(x,2) if isinstance (x,(int,float)) else x)
df_mapped1=df_mapped.applymap(lambda x: round(x,2) if isinstance (x,(int,float)) else x)
df_temp1.to_excel(writer,'Final_factors')
df_param1.to_excel(writer,'parameters')
df_mapped1.to_excel(writer,'mapped')
df_final.to_excel(writer,'Final_Analysis')
writer.save()
```
