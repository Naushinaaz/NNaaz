```python
#Importing the input CSV file
import pandas as pd
import numpy as np
import os
df1 = pd.read_csv("Zootopia_All_Plants_CustomerRatio.csv",index_col='NACS_Category')
df2 = pd.read_csv("Zootopia_All_Plants_MarginRatio1.csv",index_col='NACS_Category',usecols=['NACS_Category','Declining_cust','Num_Cust','dec_Margin_ratio'])
df1.head()
```

```python
df2.head()
```

```python
final_df= pd.merge(df1,df2,how='outer',left_index=True,right_index=True)
final_df
```

```python
#os.mkdir('Final_out')
```

```python
writer = pd.ExcelWriter('Final_out/Final_Out_Data1.xlsx')
final_df1= final_df.applymap(lambda x: round(x,2) if isinstance (x,(int,float)) else x)
final_df1.to_excel(writer,'Final_Data')
writer.save()
```
