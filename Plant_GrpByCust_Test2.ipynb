```python
import pandas as pd
import numpy as np
df = pd.read_csv("./File_for_python_Test2.csv")
```

```python
df.head()
df.fillna('X', inplace=True)
df.isnull().values.any()
```

```python
df = df.convert_objects(convert_numeric=True)
df['Sold-to Customer'] = df['Sold-to Customer'].astype(str)
df.dtypes
```

```python
df
```

```python
df['mardiff']=df['PM_2018']-df['PM_2017']
df
```

```python
len(df)
```

```python
df = df[np.isfinite(df['mardiff'])]
len(df)
```

```python
df
```

```python
margin_out=df[df['mardiff']<0].sort_values('FineLines',ascending =True)
margin_out
```

```python
final_df=margin_out.groupby(['FineLines','Sold-to Customer']).sum()
final_df
```

```python
final_df2=margin_out.groupby(['FineLines']).sum()
final_df2
```

```python
len(final_df2)
```

```python
writer = pd.ExcelWriter('Plant_Occurences_Test2_Out.xlsx')
final_df.to_excel(writer,'Sheet1')
final_df2.to_excel(writer,'Sheet2')
writer.save()
```
