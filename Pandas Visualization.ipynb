```python
idx=mcdon.index
idx
```

```python
idx=mcdon.loc['2007-01-01':'2007-05-01'].index
```

```python
idx
```

```python
stock=mcdon.loc['2007-01-01':'2007-05-01']['Adj. Close']
stock
```

```python
fig,ax = plt.subplots() #plot_date special method 
ax.plot_date(idx,stock,'-')
ax.xaxis.grid(True)
ax.yaxis.grid(True)
fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig,ax = plt.subplots() #To change the x-axis label as months using month locators
ax.plot_date(idx,stock,'-')
#LOCATING
ax.xaxis.set_major_locator(dates.MonthLocator())
#FORMATTING
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n\n%b-%Y'))

ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%a'))

fig.autofmt_xdate()
plt.tight_layout()
```
