```python
#Importing the input CSV file
import pandas as pd
import numpy as np
import os
df= pd.read_csv('Zootopia_All_plants.csv')
#cleaning the dataframe

df.head(20)
```

```python
df['Cust_ID'] = df['Cust_ID'].astype(str)
pd.options.display.float_format = '{:.2f}'.format
#df['Mar_Diff']=df['PM_2018']-df['PM_2017']
df.dtypes
```

```python
#picking valid mar_diff i.e. removing nan for mar_diff
df = df[np.isfinite(df['Mar_Diff'])]
len(df)
```

```python
#Margin_out dataframe has declining margins
margin_out=df[df['Mar_Diff']<0].sort_values('NACS_Category',ascending =True)
margin_out.head(10)
```

```python
#created df y_decline having no of customers having declining margin
final_df=margin_out.groupby(['NACS_Category'])['Cust_ID'].count()
array_keys =final_df.keys()
array_values=final_df.values
y_decline=margin_out.groupby(['NACS_Category']).sum()
y_decline['Declining_cust']=array_values
y_decline.head(20)
```

```python
#created total_cust from df where input are the only finelines from data frame y_declines
#for the fine lines , extracting total number of customers (declining + non-declining)
total_cust =df.loc[df['NACS_Category'].isin(array_keys)]
count_total_cust=total_cust.groupby(['NACS_Category'])['Cust_ID'].count()
Cust_all=count_total_cust.values
Cust_all_key=count_total_cust.keys()
```

```python
#creating a dataframe having finelines and total number of customers
ttl_Cust_df=pd.DataFrame({'Num_Cust':Cust_all},index=Cust_all_key)
ttl_Cust_df.dtypes
```

```python
#merging both the dataframes : y_decline having customers with declining margins and ttl_Cust_df having 
#total no of customers
final_df=pd.merge(y_decline,ttl_Cust_df,how='inner', left_index=True,right_index=True)
final_df['dec_Margin_ratio']=final_df['Declining_cust']/final_df['Num_Cust']
```

```python
#os.makedirs('Margin_Ratio')
```

```python
#Ypsi_All having declining margins 
#Ypsi_GreaterThan_20 having declining margins more than 20%

writer = pd.ExcelWriter('Margin_Ratio/Zootopia_All_Plants_Margin_Ratio.xlsx')
final_df1= final_df.applymap(lambda x: round(x,2) if isinstance (x,(int,float)) else x)
final_df1.to_excel(writer,'All_Plants')
final_df2=final_df1[final_df1['dec_Margin_ratio']>0.2] 
final_df2.to_excel(writer,'AllPlants_GreaterThan_20')
writer.save()
```
