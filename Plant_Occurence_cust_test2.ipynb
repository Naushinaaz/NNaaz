```python
import pandas as pd
import numpy as np
df = pd.read_csv("./File_for_python_Test2.csv")
```

```python
df.head()
df.fillna('X', inplace=True)
df.isnull().values.any()
df = df.convert_objects(convert_numeric=True)
df['Sold-to Customer'] = df['Sold-to Customer'].astype(str)
df.dtypesdf['mardiff']=df['PM_2018']-df['PM_2017']
df
```

```python
df = df[np.isfinite(df['mardiff'])]
len(df)
```

```python
margin_out=df[df['mardiff']<0].sort_values('FineLines',ascending =True)
margin_out
```

```python
final_df=margin_out.groupby(['FineLines'])['Sold-to Customer'].count()
array_keys =final_df.keys()
array_values=final_df.values
y_decline=margin_out.groupby(['FineLines']).sum()
y_decline['few_cust']=array_values

```

```python
all_cust =df.loc[df['FineLines'].isin(array_keys)]
count_all=all_cust.groupby(['FineLines'])['Sold-to Customer'].count()
Cust_all=count_all.values
cust_all_key=count_all.keys()
```

```python
nn=pd.DataFrame({'No_Cust':Cust_all},index=cust_all_key)
nn.dtypes
```

```python
final_nn=pd.merge(y_decline,nn,how='inner', left_index=True,right_index=True)
final_nn['ratio']=final_nn['few_cust']/final_nn['No_Cust']
writer = pd.ExcelWriter('MostWantedColumn.xlsx')
final_nn.to_excel(writer,'Sheet1')
writer.save()
```
