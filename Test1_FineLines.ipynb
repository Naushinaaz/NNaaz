```python
import pandas as pd
import numpy as np
df = pd.read_csv("./Analysis_Finelines1.csv")
```

```python
df.groupby(["Customers"]).sum().sort_values("Total Sales", ascending=False)
```

```python
df["Profit"] = df["Customer Net Price"] - df["BAC - Best Acq Cost"]
df["ProfitPerUnit"]=df["Profit"]/df["Blended Qty-Sales UOM"]
df["ProfitMargin"]=df["Profit"]/df["Customer Net Price"]
df.head()
```

```python
df.groupby("Customers").sum().sort_values("ProfitMargin", ascending=False).head()
```

```python
group = df.groupby(["FineLines","Customers"]).sum()
profit_margin = group["ProfitMargin"].groupby(level=0, group_keys=False)
profit_margin.nlargest(5)
```

```python
cus_plot = df.groupby(["FineLines","Customers"]).sum().plot(kind='bar')
cus_plot.set_xlabel("Customers")
cus_plot.set_ylabel("ProfitMargin")
```
