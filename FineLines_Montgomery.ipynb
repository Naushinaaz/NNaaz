```python
df_monty = df_monty.convert_objects(convert_numeric=True)
```

```python
df_monty.groupby(['FineLines']).sum().sort_values('PM_2017', ascending=False)
```

```python
df_monty['MarDiff']=df_monty['PM_2018']-df_monty['PM_2017']
```

```python
MD_finelines=df_monty.groupby(["FineLines"]).sum().sort_values("MarDiff", ascending=True)
```

```python
MD_finelines
```

```python
margin_out=MD_finelines[MD_finelines['MarDiff']<0].sort_values('MarDiff',ascending =True)
```
